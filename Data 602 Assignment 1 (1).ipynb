Create nodes and edges for a basic graph, a as the Krackhardt kite.


Since we are going to create a network graph, we need to include a python package such as networkx.
Since we also want to show a visual of the nodes, edges and graphs that we create, we also include the presentation package matplotlib

```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Krackhardt_kite = nx.Graph()
```

Define the Krackhardt kite nodes using a Python list

```python
knodes = ['Andre', 'Beverly', 'Carol', 'Diane', 'Ed', 'Fernando', 'Garth', 'Heather', 'Ike', 'Jane']
```

```python
Add the nodes to the graph using the add node function of networkx
```

```python
for name in knodes:
    Krackhardt_kite.add_node(name)
```

Show all the nodes visually; make sure no names are missing (10 nodes)

```python
nx.draw(Krackhardt_kite,with_labels=1)
plt.show()
```

Define the edges of the graph (this identifies it as a Krackhardt kite)

```python
kedges = [('Andre','Beverly'), ('Andre', 'Carol'), ('Andre', 'Diane'), ('Andre', 'Fernando'), ('Beverly', 'Diane'), 
          ('Beverly', 'Ed' ), ('Beverly', 'Garth'), ('Carol', 'Diane'), ('Carol', 'Fernando'),  ('Diane', 'Ed'), 
          ('Diane', 'Fernando'), ('Diane', 'Garth'), ('Ed', 'Garth'), ('Fernando', 'Garth'), ('Fernando', 'Heather'),  
          ('Garth', 'Heather'), ('Heather', 'Ike'), ('Ike', 'Jane')]
```

Add all the defined edges to the existing graph (Krackhardt kite)

```python
Krackhardt_kite.add_edges_from(kedges)
```

Display the graph for a visualization of the interactions amongst all the nodes; note that different nodes have different levels of interaction and centrality.

```python
nx.draw(Krackhardt_kite,with_labels=1)
plt.show()
```
